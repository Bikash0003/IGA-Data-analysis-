```python
# High-risk: low sales (Q1) + high discounts (Q4)
low_sales = df['Amount inc'] < df['Amount inc'].quantile(0.25)
high_discounts = df['Specials % share'] > df['Specials % share'].quantile(0.75)
high_risk = df[low_sales & high_discounts]

# Optimization: high sales (Q4) + low GP% (Q1)
high_sales = df['Amount inc'] > df['Amount inc'].quantile(0.75)
low_gp = df['GP%'] < df['GP%'].quantile(0.25)
optimize_opportunity = df[high_sales & low_gp]
result = pd.concat([high_risk, optimize_opportunity])
print("High-Risk and Optimization Departments")
print(result)

```

```python
df2 = pd.read_excel("/content/Austrilian_supermarkets.xlsx")
df2.head()
```

```python
df2_pivot = df2.pivot_table(
    index='Supermarket',
    columns='Category',
    values='Amount',
    aggfunc='first'
).reset_index()

# Step 3: Clean columns and convert to numeric
df2_pivot['Revenue'] = df2_pivot['Revenue'].str.extract(r'(\d+\.?\d*)')[0].astype(float)
df2_pivot['Waste Diversion'] = df2_pivot['Waste Diversion'].str.replace('%', '', regex=False).astype(float)
df2_pivot['Emissions Reduction'] = df2_pivot['Emissions Reduction'].str.replace('%', '', regex=False).astype(float)

# Step 4: Filter out Aldi and Spudshed
df_plot_filtered = df2_pivot[~df2_pivot['Supermarket'].isin(['Aldi', 'Spudshed'])]

# Step 5: Plot comparisons
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Sustainability & Financial Comparison of Major Supermarkets (Excluding Aldi & Spudshed)', fontsize=14)

# Revenue
axes[0].bar(df_plot_filtered['Supermarket'], df_plot_filtered['Revenue'], color='skyblue')
axes[0].set_title('Revenue ($ Billion)')
axes[0].set_ylabel('Billions $')

# Waste Diversion
axes[1].bar(df_plot_filtered['Supermarket'], df_plot_filtered['Waste Diversion'], color='lightgreen')
axes[1].set_title('Waste Diversion (%)')
axes[1].set_ylabel('% Diverted')

# Emissions Reduction
axes[2].bar(df_plot_filtered['Supermarket'], df_plot_filtered['Emissions Reduction'], color='salmon')
axes[2].set_title('Emissions Reduction (%)')
axes[2].set_ylabel('% Reduced')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
